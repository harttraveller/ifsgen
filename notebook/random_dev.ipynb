```python
def viz(a, b, c):
    fig = go.Figure(
        data=[
            go.Scatter3d(
                x=a,
                y=b,
                z=c,
                mode="markers",
                marker=dict(
                    size=1,
                    color="white",
                    showscale=False,
                    reversescale=True,
                ),
                hoverinfo="skip",
            )
        ]
    )
    from ifs.minimal import minimal_template

    fig.update_layout(template=minimal_template)
    fig.update_layout(
        autosize=False,
        width=480,
        height=480,
        margin=dict(l=0, r=0, b=0, t=0, pad=0),
        yaxis_zeroline=False,
        xaxis_zeroline=False,
        paper_bgcolor="rgba(0,0,0,0)",
        plot_bgcolor="rgba(0,0,0,0)",
    )
    fig.update_yaxes(title="asdf")
    camera = dict(
        up=dict(x=0, y=0, z=1), center=dict(x=-0, y=0, z=0), eye=dict(x=1, y=1, z=1)
    )
    fig.update_layout(scene_camera=camera)
    fig.show(config=dict(displayModeBar=False))
```

```python
points = sierpinski(20000,3)
points = [list(i) for i in points]
points = sorted(points, key = lambda x: x[-1])
# points = [np.dot(rotation_matrix(axis, theta), i) for i in points]
a, b, c = zip(*points)
a = np.array(a)
b = np.array(b)
c = np.array(c)
viz(a,b,c)
```
